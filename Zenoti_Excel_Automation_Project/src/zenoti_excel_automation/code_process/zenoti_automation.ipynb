```python
# Import necessary libraries
import os
from datetime import datetime as dt
from pathlib import Path
import polars as pl
import requests
from dotenv import load_dotenv
import src.zenoti_excel_automation as zea
from src.zenoti_excel_automation import assets
```

```python
# Get cwd
cwd = os.getcwd()
parent_dir = Path(cwd).parent
config_path = os.path.join(parent_dir, "assets", "config.env")

# Load environment variables from .env file
load_dotenv(dotenv_path=config_path)
```

```python
# Function to connect to Zenoti API
def connect_to_zenoti(api_key, url):
    """
    Connect to Zenoti API using the provided API key.
    """
    headers = {"accept": "application/json", "Authorization": f"apikey {api_key}"}

    response = requests.get(url, headers=headers)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        try:
            # Attempt to parse and return the JSON response
            return response.json()
        except requests.exceptions.JSONDecodeError:
            # Handle cases where the response is successful but not JSON
            print("Error: The response was not in a valid JSON format.")
            print("Response text:", response.text)
            return None
    else:
        # Handle unsuccessful requests
        print(f"Failed to connect to Zenoti API. Status code: {response.status_code}")
        print("Response text:", response.text)
        return None
```

```python
# Set up api key and url
zenoti_api_key = os.getenv("ZENOTI_API_KEY")
zenoti_url = os.getenv("ZENOTI_TEST_URL")

# Test connection to Zenoti API
response_data = connect_to_zenoti(zenoti_api_key, zenoti_url)

if response_data:
    print("Successfully connected to Zenoti API and received data.")
    # You can then proceed to work with the response_data
else:
    print("Failed to get a valid response from Zenoti API.")
```
