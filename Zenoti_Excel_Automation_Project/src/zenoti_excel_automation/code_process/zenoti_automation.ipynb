```python
# Import the reusable Zenoti API helpers from the utils module
from zenoti_excel_automation.code_process.utils import load_config, call_zenoti_api
```

```python
# Load API credentials and center GUIDs from config.env
config = load_config()
```

```python
# Verify connectivity by listing all configured centers
result = call_zenoti_api(config, "/centers")
if result:
    for center in result.get("centers", []):
        print(f'{center["id"]}  {center["code"]:15s}  {center["name"]}')
else:
    print("Failed to list centers.")
```

```python
# Fetch stock movement data for each center
for center_id, center_name in config["centers"].items():
    if not center_id:
        print(f"{center_name}: no GUID set in config.env, skipping")
        continue
    params = {
        "center_id": center_id,
        "start_date": "2025-05-01",
        "end_date": "2025-05-31",
        "report_type": 1,
    }
    result = call_zenoti_api(config, "/inventory/stock_movement", params=params)
    if result:
        items = result.get("list", [])
        print(f"{center_name}: {len(items)} records")
        for item in items[:2]:
            print(f'  {item["product_code"]}  {item["product_name"][:50]:50s}  store: {item["ending_store"]}  floor: {item["ending_floor"]}')
        if len(items) > 2:
            print(f'  ... and {len(items) - 2} more')
    print()
```
